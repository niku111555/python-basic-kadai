```python
user_names = ["侍太郎", "侍一郎", "侍次郎", "侍三郎", "侍四郎"]
print(user_names[1]) # 2番目の要素だけ表示

user_names[1] = "侍花子" # 2番目の要素を更新
print(user_names)

user_names.append("侍五郎") # 6番目の要素に追加
print(user_names)

user_names.pop(2) # 3番目の要素を削除(0始まりのため2を指定)
print(user_names)

```

```python
weekly_weather = ["月曜日は晴れです","火曜日は雨です","水曜日は晴れです","木曜日は晴れです","金曜日は曇りです","土曜日は曇りのち雨です","日曜日は雷雨です"]
print(weekly_weather[2]) # 3番目の要素だけ表示

weekly_weather = {"mon":"晴れ","tue":"雨","wed":"晴れ","thu":"晴れ","fri":"曇り","sat":"曇りのち雨","sun":"雷雨"}
 # キー「wed」の値を表示
print(weekly_weather["wed"])
```

```python
# 連想配列を作成
personal_data = {"name":"侍太郎","age":36,"gender":"男性"}

# キー「name」の値を表示
print(personal_data["name"])

# 連想配列全体を表示
print(personal_data)

# キー「age」の値を更新
personal_data["age"] = 37

# 新しく要素（キーと値）を追加
personal_data["address"] = "東京都"

# 連想配列全体を表示
print(personal_data)
```

```python
country_names = ("日本","アメリカ","イギリス","フランス")
# 3番目の要素を取り出す
print(country_names[2])

# 全ての要素を取り出す
print(country_names)
```

```python
country_names = {"アメリカ","イギリス","日本","フランス"}
# セット全体を表示
print(country_names)

country_names.add("ドイツ") # 「ドイツ」をセットに追加
print(country_names)

country_names.remove("イギリス") # セットから「イギリス」を削除
print(country_names)

```
